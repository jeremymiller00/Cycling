```python
import sys
sys.path.append('..')
import json
```

```python
from __future__ import print_function
```

```python
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
with open('../data/sample_data.json', 'r') as f:
    data = json.load(f)
```

```python
texts = []
for article in data['webscrape']:
    texts.append(article['text'])
```

```python
len(texts)
```

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5, 
                                min_df = 2)
dtm_tf = tf_vectorizer.fit_transform(texts)
print(dtm_tf.shape)
```

```python
type(dtm_tf)
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(texts)
print(dtm_tfidf.shape)
```

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_topics=20, random_state=0)
lda_tf.fit(dtm_tf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_topics=20, random_state=0)
lda_tfidf.fit(dtm_tfidf)
```

```python
prepared_data = pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer)
```

```python
pyLDAvis.display(prepared_data)
```

```python
pyLDAvis.save_html(prepared_data, 'lda.html')
```

```python
pyLDAvis.sklearn.prepare(lda_tfidf, dtm_tfidf, tfidf_vectorizer)
```
